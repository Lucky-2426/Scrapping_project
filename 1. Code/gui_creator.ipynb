```python
import sys

from PyQt5.QtWidgets import (
    QApplication,
    QHBoxLayout,
    QVBoxLayout,
    QPushButton,
    QWidget,
    QScrollArea,
    QLabel,
    QComboBox,
)

import os
from PyQt5 import QtCore
import json
```

```python
class Window(QWidget):
    def __init__(self):
        super(Window, self).__init__()
        self.load_config_files()
        self.setWindowTitle('Pimp my Scrap (PMS)')
        layout = QHBoxLayout()

        self.button_left  = QPushButton('Left Button')
        self.button_right = QPushButton('Right Button')
        self.widget       = QWidget()
        self.center       = QScrollArea()
        self.vbox         = QVBoxLayout()
        
        for task in self.page_elements:
            vertical_container = QVBoxLayout()
            task_name = QLabel(task)
            task_name.setObjectName("title")
            vertical_container.addWidget(task_name)
            print('Task Name: ', task)
            for action_name in self.page_elements[task]:
                action_box = self.get_tasks_box(task, action_name)
                vertical_container.addLayout(action_box)

            self.vbox.addLayout(vertical_container)

        self.widget.setLayout(self.vbox)
        self.center.setWidget(self.widget)

        # Scroll Area Properties
        self.center.setVerticalScrollBarPolicy(QtCore.Qt.ScrollBarAlwaysOn)
        self.center.setHorizontalScrollBarPolicy(QtCore.Qt.ScrollBarAlwaysOn)
        self.center.setWidgetResizable(True)

        layout.addWidget(self.button_left, 1)
        layout.addWidget(self.center, 3)
        layout.addWidget(self.button_right, 1)

        # self.button_left.clicked.connect(self.load_tasks)

        self.setLayout(layout)

        style = """
                QLabel{
                    color: #E9E9E9;
                    font-size: 24px;
                    background: #0C2637;
                    border-radius: 15px;
                    font-family: Calibri;
                    padding: 2px 15px;
                }

                QLabel#title{
                    color: #0274B3;
                    font-size: 32px;
                    background: #F2F2F2;
                    border-radius: 15px;
                    font-family: Calibri;
                    padding: 2px 15px;
                }

                QLineEdit{
                    color: #0274B3;
                    background: #F2F2F2;
                    border-radius: 15px;
                    padding: 2px 15px;
                    font-size: 16px;
                }

                QWindow{
                    background: #FFFFFF;
                }
        """

        self.setStyleSheet(style)
    
    def get_tasks_box(self, task, action):
        # Create a Horizontal Box Layout to contain the action info
        hbox        = QHBoxLayout()

        # Creating a combobox with all the possible actions
        to_do       = self.page_elements[task][action]['to_do']
        to_do_combo = QComboBox()
        to_do_combo.addItems(['Fill', 'Click', 'Wait', 'Load', 'Sleep', 'Get', 'Export', 'Scrap'])
        to_do_combo.setCurrentText(to_do)
        hbox.addWidget(to_do_combo, 1)

        # Unpacking the params to create a new layout with the params
        params = self.page_elements[task][action]['params']

        for elem in params:
            # Create a Label for the element's name
            label_element = QLabel(elem)
            label_element.setObjectName('text_element')

            # 
            combo_element = QComboBox()
            combo_element.addItems(self.page_fields)
            combo_element.setCurrentText(str(params[elem]))
            combo_element.setObjectName('combo_element')

            hbox.addWidget(label_element, 1)
            hbox.addWidget(combo_element, 1)

        return hbox

    def load_config_files(self):
        # Opening the tasks.json file to load the page elements
        file_task = open(os.path.join('..', '7. Config', 'tasks.json'))
        self.page_elements = json.load(file_task)['LinkedIn']

        file_login = open(os.path.join('..', '7. Config', 'login.json'))
        self.web_elements = json.load(file_login)['page_elements']

        self.page_fields = []
        for page_field in self.web_elements:
            self.page_fields.append(page_field)
```
